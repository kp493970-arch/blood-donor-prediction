```python
# Cell 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, roc_curve, ConfusionMatrixDisplay
)

print("All libraries loaded successfully")
```

```python
# Cell 2: Load the data
df = pd.read_csv('../data/transfusion.csv')

# Rename columns to clean short names
df.columns = ['Recency', 'Frequency', 'Monetary', 'Time', 'Class']

print("Dataset loaded successfully")
print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Cell 3: Explore the data (EDA)
print("=== CLASS BALANCE ===")
print(df['Class'].value_counts())
print(f"\nPositive rate: {df['Class'].mean()*100:.1f}%")
print("\nThis is the imbalance problem — only 24% donated again")
print("If model predicts NO for everyone = 76% accuracy but completely useless")
print("So we use AUC instead of accuracy as our main metric")
```

```python
# Cell 4: Feature Engineering
# Monetary = Frequency x 250 exactly — perfectly redundant
# Keeping it gives the model duplicate information (multicollinearity)
# We drop it deliberately — this is a conscious modeling decision

print("Checking multicollinearity:")
print(df[['Frequency', 'Monetary']].head(10))
print("\nMonetary is exactly Frequency x 250 — confirmed redundant")

df_clean = df.drop(columns=['Monetary'])

X = df_clean.drop(columns=['Class'])
y = df_clean['Class']

print(f"\nFeatures we will use: {list(X.columns)}")
print("Dropped: Monetary")
```

```python
# Cell 5: Train/Test Split
# 75% of data trains the model, 25% tests it on unseen donors
# stratify=y keeps same 24% positive rate in both train and test
# random_state=42 makes the split reproducible every time you run it

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print(f"Training set: {len(X_train)} donors")
print(f"Test set:     {len(X_test)} donors")
print(f"\nTrain positive rate: {y_train.mean()*100:.1f}%")
print(f"Test positive rate:  {y_test.mean()*100:.1f}%")
print("\nBoth rates match — stratification worked correctly")
```

```python
# Cell 6: Scaling
# Recency ranges 0-74, Time ranges 2-98 — very different scales
# Logistic regression treats bigger numbers as more important
# StandardScaler fixes this by converting everything to same scale
# CRITICAL: fit scaler on training data only
# Fitting on all data = test information leaks into training = data leakage
# Data leakage = falsely optimistic performance = dishonest results

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # learn scale from train only
X_test_scaled  = scaler.transform(X_test)        # apply same scale to test

print("Scaling complete")
print(f"\nOriginal Recency range: {X['Recency'].min()} to {X['Recency'].max()}")
print(f"Scaled Recency range:   {X_train_scaled[:,0].min():.2f} to {X_train_scaled[:,0].max():.2f}")
print("\nAll features now on same scale — model treats them fairly")
```

```python
# Cell 7: Train the model
# class_weight='balanced' handles the 24% imbalance
# It tells the model: missing a real donor is more costly than a false alarm
# Without this, model would predict NO for almost everyone and look 76% accurate
# max_iter=1000 gives the model enough steps to find the best coefficients

model = LogisticRegression(class_weight='balanced', random_state=42, max_iter=1000)
model.fit(X_train_scaled, y_train)

print("Model trained successfully on 561 donors")
print("\nCoefficients (direction of each feature's effect):")
coef_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_[0]
})
print(coef_df.to_string(index=False))
print("\nNegative Recency = more recent donors are MORE likely to return")
print("Positive Frequency = donors who gave more times are MORE likely to return")
```

```python
# Cell 8: Evaluate the model on 187 unseen test donors
y_pred = model.predict(X_test_scaled)
y_pred_prob = model.predict_proba(X_test_scaled)[:, 1]

print("=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred, target_names=['No Donation', 'Donation']))

print("=== CONFUSION MATRIX ===")
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("""
Layout:
[[TN  FP]
 [FN  TP]]

TN = non-donors correctly cleared
FP = false alarms (wrongly flagged as donors)
FN = real donors missed  <- most costly
TP = real donors correctly caught
""")

auc = roc_auc_score(y_test, y_pred_prob)
print(f"ROC-AUC: {auc:.3f}")
print("0.5 = coin flip | 0.785 = genuinely useful model")
```

```python
# Cell 9: ROC Curve
# Shows model performance across every possible threshold
# X axis: False Positive Rate (false alarms)
# Y axis: True Positive Rate (donors correctly caught)
# Dotted diagonal = random coin flip (AUC 0.5)
# Our curve hugs top-left = better than random

fpr, tpr, _ = roc_curve(y_test, y_pred_prob)

plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, color='steelblue', lw=2,
         label=f'Logistic Regression (AUC = {auc:.3f})')
plt.plot([0, 1], [0, 1], color='gray', lw=1,
         linestyle='--', label='Random baseline (AUC = 0.5)')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.title('ROC Curve: Blood Donor Return Prediction')
plt.legend()
plt.tight_layout()
plt.show()
```
